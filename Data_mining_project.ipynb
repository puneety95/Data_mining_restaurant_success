```python
df=df.drop(['Web_Link' , 'Phone_number','Detail_address','spam_review','5_star_review_percentage','4_star_review_percentage','3_star_review_percentage',	'2_star_review_percentage',	'1_star_review_percentage'], axis=1)
print(df.head())
```

```python
count=0
for i in df.columns:
  if i=="Wine and Beer":
    break;
  count=count+1


df2 = pd.DataFrame(columns = ['Service_Type', 'No_count', 'Yes_count','%No' , '%Yes'])

                   
for i in (df.columns[count:]):
  ser=df[i].value_counts()
  df2=df2.append({'Service_Type':ser.name, 'No_count': ser[0] , 'Yes_count':ser[1] , '%No': (ser[0]/(ser[0]+ser[1]))*100 , '%Yes': (ser[1]/(ser[0]+ser[1]))*100}, ignore_index=True)

for i in (df2.loc[df2['%No']>90]).index:
  df=df.drop([df2['Service_Type'][i]], axis=1)
  


```

```python
print(df.shape)
```

```python
df.head()
```

```python
#mapping values of payment modes from ordinal to digits
pay_modes = df['payment_modes'].unique()
new_val=1
pay_modes_values = pd.DataFrame(columns = ['Mode', 'New_value'])

for i in pay_modes:
  pay_modes_values=pay_modes_values.append({'Mode':i, 'New_value': new_val}, ignore_index=True)
  df['payment_modes']=df['payment_modes'].replace(to_replace=i , value=new_val)
  new_val=new_val+1
```
